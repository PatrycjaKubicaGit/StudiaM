```python
import numpy as np
import matplotlib.pyplot as plt

# Function to compute the IDFT matrix
def idft_matrix(N):
    """Compute the IDFT matrix."""
    k = np.arange(N)  # Sample indices
    mu = k.reshape((N, 1))  # Frequency indices
    W = np.exp(1j * 2 * np.pi * mu * k / N) / N  # IDFT matrix
    return W

# Defining the input signal x
x_mu = np.array([6, 4, 4, 5, 3, 4, 5, 0, 0, 0, 0], dtype=complex)

# Trying different values of N
N_values = [4, 8, 11]
for N in N_values:
    # Computing the IDFT matrix
    W = idft_matrix(N)
    
    # Computing the synthesized signal
    x_synthesized = W @ x_mu[:N]
    
    # Plotting the results
    plt.figure(figsize=(8, 4))
    plt.plot(np.real(x_synthesized), 'C0o-', lw=0.5, label='Real part')  
    plt.plot(np.imag(x_synthesized), 'C1o-', lw=0.5, label='Imaginary part')  
    plt.xlabel('Sample index')
    plt.ylabel('Amplitude')
    plt.title(f'Synthesized signal using IDFT (N={N})')
    plt.legend()
    plt.grid()
    plt.show()
    
    # Printing the W and K matrices (DFT matrix, inverse of W)
    print(f'IDFT matrix W for N={N}:\n', W)
    print(f'DFT matrix K for N={N}:\n', np.conj(W).T * N)

```
